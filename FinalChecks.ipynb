```python
title = input('Enter title: ')
art = df.loc[df['Article Title'].str.contains(title)]
art
```

```python
auths = art.iloc[0]['Author(s)'].split('.,')
```

```python
auths_first = auths[:7]
auths_last = auths[-1]
```

```python
len(auths)
```

```python
auths_first
```

```python
auths_last
```

```python
auth_m = {i : sorted(auths)[i] for i in range(len(sorted(auths)))}
auth_m
```

```python
l = []
while True:
  try:
    num = int(input("Enter number next to name ('Just click ENTER to end'): "))
    l.append(num)
  except:
    break
```

```python
incl = [auth_m[num] for num in l]
UF_affil = ""
for i in range(len(incl)):
  if incl[i] not in auths_first:
    UF_affil += incl[i]
    UF_affil += ".," if i < len(incl)-1 else "."

UF_affil
```

```python
s = ""
for a in auths_first:
  s += a
  s += '.,'

s += (f" (+ {len(auths) - 8} authors{' including' if len(UF_affil) > 0 else ''}{UF_affil}),")
s += auths_last
```
