Formatting the author names when manually entering a citation in Endnote is the most time consuming part. Use this program to complete this task for you.

Steps:
1. Run the cell below (running  cell is accomplished by pressing the play button on the top left of the cell when hovering over it with your mouse)
2. Choose separator (a separator is what separates the names from each other. For example, in the list of authors [Salisbury, Allyson B.; Miesbauer, Jason W.; Koeser, Andrew K.], the seaprator is the semi-colon, as it separates each full name)
3. Run the cell based on which format the list is in (are the last names listed first, or are the first names listed first)?
4. Enter the list of authors in the textbox that pops up under the cell
5. Copy the output into Endnote

***Alway double check that the output is correct!!!***

Special note: You can enter the full recommended citation for Journal of Agricultural Education articles in the last cell to extract all the information in a readable format.

```python
#RUN THIS CELL FIRST!!!
def authors(str):
  names = str.split(", ")
  #print(names)
  new_str = ""
  for x in range(len(names)):
    n = names[x]
    n = n.replace("and ", "")
    n = n.replace("& ", "")
    if x % 2 == 0:
      new_str += f"{n},"
    else:
      new_str += f" {n}"
      new_str += "\n"

  print(new_str)
```

### Separator: Semicolon

```python
#Format: Last, First, Middle
#Example: Salisbury, Allyson B.; Miesbauer, Jason W.; Koeser, Andrew K.
str = input("Enter string:\n")
str = str.replace("And ", "")
str = str.replace("and ", "")
str = str.replace("& ", "")
str = str.replace('; ', '\n')
print("Output:\n")
print(str)
```

```python
#Format: First, Middle, Last
#Example: "Allyson B. Salisbury; Jason W. Miesbauer; and Andrew K. Koeser"
str = input("Enter list of authors:\n")
names = str.split("; ")
new_str = ""
print("Output:\n")
for n in names:
  n = n.replace("And ", "")
  n = n.replace("and ", "")
  n = n.replace("and ", "")
  j = n.split(" ")
  new_str += f"{j[-1]},"
  for i in range(len(j)-1):
    new_str += (f" {j[i]}")
  new_str += "\n"

print(new_str)
```

### Separator: Comma

```python
#Format: First, Middle, Last
#Example: "Allyson B. Salisbury, Jason W. Miesbauer, and Andrew K. Koeser"
str = input("Enter list of authors:\n")
names = str.split(", ")
new_str = ""
print("Output:\n")
for n in names:
  n = n.replace("And ", "")
  n = n.replace("and ", "")
  n = n.replace("and ", "")
  j = n.split(" ")
  new_str += f"{j[-1]},"
  for i in range(len(j)-1):
    new_str += (f" {j[i]}")
  new_str += "\n"

print(new_str)
```

```python
#Format: Last, First, Middle
#Example: Christopher Wynveen, Ingrid Schneider, Deonne VanderWoude, Taylor Stein, Heather Gibson, Kim Shinew, William Hendricks, and Megha Budruk
str = input("Enter authors:")
print("Output:\n")
authors(str)
```

### Format Full Citations

```python
#Enter full Journal of Agricultural Education Recommended Citation
citation = input("ENTER CITATION: \n")
print("Output:")
print("\n")
auths = citation.split("(")[0]
authors(auths)
title = citation.split('). ')[1].split(".")[0]
print(title)
year = citation[(citation.find("(") + 1):citation.find("(") + 5]
print("Y:", year)
journal = citation.split(",")[-3].split(". ")[-1]
print("J:", journal)
vol_iss = citation.split(", ")[-2]
vol = vol_iss[:vol_iss.find("(")]
iss = vol_iss[(vol_iss.find("(")+1):-1]
print("V:", vol)
print("I:", iss)
pages = citation.split(". ")[-2].split(", ")[-1]
print("P:", pages)
doi = citation.split(" ")[-1][16:]
print("DOI:", doi)
```
